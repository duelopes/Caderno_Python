## Teste de hiptese

**H<sup>null<sup>**
>**A distribuio da taxa da gorjeta  a mesma nos dois grupos**

**H<sup>alt<sup>**
>**A distribuio da taxa da gorjeta no  a mesma nos dois grupos**

```python
from scipy.stats import ranksums
```

```python
sobremesa = gorjetas.query("sobremesa == 'Sim'").porcentagem
```

```python
sem_sobremesa = gorjetas.query("sobremesa == 'No'").porcentagem
```

```python
r = ranksums(sobremesa, sem_sobremesa)
```

```python
print('O valor do p-value  {}'.format(r.pvalue))
```

**H<sup>null<sup>**
>**A distribuio da taxa da gorjeta  a mesma nos dois grupos**
